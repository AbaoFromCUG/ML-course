1.引入必要的包。

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

2.加载数据，浏览数据集，并对数据进行可视化。

```python
df = pd.read_csv("https://raw.githubusercontent.com/qikunlun/ML-course/master/data/testSet-KMeans.csv", header=None)
plt.scatter(df.values[:,0], df.values[:,1])
plt.show()
```

4.采用KMeans对点进行聚类并可视化。 

```python
plt.subplot(2, 2, 1)
kmeans = KMeans(n_clusters=2, random_state=0).fit(df.values)
plt.scatter(df.values[:,0], df.values[:,1], c=kmeans.labels_)
plt.subplot(2, 2, 2)
kmeans = KMeans(n_clusters=3, random_state=0).fit(df.values)
plt.scatter(df.values[:,0], df.values[:,1], c=kmeans.labels_)
plt.subplot(2, 2, 3)
kmeans = KMeans(n_clusters=4, random_state=0).fit(df.values)
plt.scatter(df.values[:,0], df.values[:,1], c=kmeans.labels_)
plt.subplot(2, 2, 4)
kmeans = KMeans(n_clusters=5, random_state=0).fit(df.values)
plt.scatter(df.values[:,0], df.values[:,1], c=kmeans.labels_)
plt.show()
```
